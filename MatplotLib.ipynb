```python

import numpy as np
```

```python
# Create Data
x = np.linspace(0, 10, 100)
x
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Plot from  a Dictionary
butter = {"Almond": 10,
         "Peanut": 8,
         "Cashew": 12}
fig, ax = plt.subplots()
ax.bar(butter.keys(), butter.values())
ax.set(title="Butter Shop")
```

```python
fig, ax = plt.subplots()
ax.barh(list(butter.keys()), list(butter.values()));
```

```python
# Histogram
# Prepare Data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two Options for subplots

```python
# Option-1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(butter.keys(), butter.values());
ax4.hist(np.random.randn(1000));
```

```python
# Option-2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(butter.keys(), butter.values());
ax[1, 1].hist(np.random.randn(1000));
```

```python
# Heart Disease
heart_disease = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/heart-disease.csv")
heart_disease
```

```python
heart_disease["age"].plot.hist(bins=10);
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10, 30), subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
# ax.set_xlim([45, 100]);
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

ax.set(title="Heart Disease and Cholesterol Level",
       xlabel="age",
       ylabel="Cholesterol")

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

## Plots

```python
# For the figure
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# For axes 0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

ax0.set(title="Heart Disease and Cholesterol Level",
       ylabel="Cholesterol")

ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(over_50["chol"].mean(),
           linestyle="--");

# For axes 1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="age",
       ylabel="Thalach")

ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(),
           linestyle="--")

fig.suptitle("Heart Analysis");
```

```python
# Set Style
plt.style.use("seaborn-whitegrid")
# For the figure
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# For axes 0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                      cmap="winter")

ax0.set(title="Heart Disease and Cholesterol Level",
       ylabel="Cholesterol")
ax0.set_xlim([50, 80])

ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(over_50["chol"].mean(),
           linestyle="--");

# For axes 1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                      cmap="winter")

ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="age",
       ylabel="Thalach")
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(),
           linestyle="--")

fig.suptitle("Heart Analysis");
```
