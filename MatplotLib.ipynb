##  Making figure with Numpy Array
* Linear,
* Scatter,
* Bar Plot and
* Histogram

```python
import numpy as np
```

```python
# Create Data
x = np.linspace(0, 10, 100)
x
```

```python
# Plot Data
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Use this data to create a scatterplot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Plot from  a Dictionary
butter = {"Almond": 10,
         "Peanut": 8,
         "Cashew": 12}
fig, ax = plt.subplots()
ax.bar(butter.keys(), butter.values())
ax.set(title="Butter Shop")
```

```python
fig, ax = plt.subplots()
ax.barh(list(butter.keys()), list(butter.values()));
```

```python
# Histogram
# Prepare Data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two Options for subplots

```python
# Option-1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(butter.keys(), butter.values());
ax4.hist(np.random.randn(1000));
```

```python
# Option-2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(butter.keys(), butter.values());
ax[1, 1].hist(np.random.randn(1000));
```
